### Imports

```python
import pandas as pd
```

### Read data

```python
# Read 'xlsx' files as dataframes
chuquisaca = pd.read_excel('../data/raw/feminicidios_bolivia.xlsx', sheet_name='Chuquisaca')
cochabamba = pd.read_excel('../data/raw/feminicidios_bolivia.xlsx', sheet_name='Cochabamba')
santa_cruz = pd.read_excel('../data/raw/feminicidios_bolivia.xlsx', sheet_name='Santa Cruz')
potosi = pd.read_excel('../data/raw/feminicidios_bolivia.xlsx', sheet_name='Potosí')
oruro = pd.read_excel('../data/raw/feminicidios_bolivia.xlsx', sheet_name='Oruro')
tarija = pd.read_excel('../data/raw/feminicidios_bolivia.xlsx', sheet_name='Tarija')
la_paz = pd.read_excel('../data/raw/feminicidios_bolivia.xlsx', sheet_name='La Paz')
beni = pd.read_excel('../data/raw/feminicidios_bolivia.xlsx', sheet_name='Beni')
pando = pd.read_excel('../data/raw/feminicidios_bolivia.xlsx', sheet_name='Pando')
```

### Process data

```python
# Concatenate dataframes
df_join = pd.concat([chuquisaca, cochabamba, santa_cruz, potosi, oruro, tarija, la_paz, beni, pando], axis=0)
df_join
```

```python
# Convert columns into a list
cols = df_join.columns
```

```python
# Remove new lines
for i in range(len(cols)):
    df_join[cols[i]] = df_join[cols[i]].replace(r'\n*', '', regex=True)
# Show dataframe
df_join
```

### Save data

```python
# Save dataframe as 'xlsx'
df_join.to_excel('../data/processed/feminicidios_bolivia_procesada.xlsx', index=False)
```
